```python
from keras.preprocessing import image
from keras.layers import *
from keras.applications.resnet50 import ResNet50
from keras.models import Model
import numpy as np
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
```

```python
generator = image.ImageDataGenerator(horizontal_flip=True,rotation_range=10, shear_range=.2,zoom_range=.5)
```

```python
gen = generator.flow_from_directory("../input/indian_notes/indian_notes/",target_size=(700,700), batch_size=1)
```

```python
images, values = gen.next()
```

```python
resnet_mod = ResNet50(include_top=False, weights='imagenet', input_shape=(700, 700, 3), classes=4)
```

```python
flat = Flatten()(resnet_mod.output)
d1 = Dense(10000, activation="tanh") (flat)
d2 = Dense(5000, activation="tanh") (d1)
dr1 = Dropout(rate=0.4)(d2)
d3 = Dense(1000, activation="tanh") (dr1)
d4 = Dense(500, activation="tanh") (d3)
d5 = Dense(100, activation="tanh") (d4)
dr = Dropout(rate=0.4)(d5)
d6 = Dense(50, activation="tanh") (dr)
out = Dense(4, activation="softmax") (d3)
```

```python
c=133
X=[]
y=[]
while(c>0):
    images,values=gen.next()
    for img,val in zip(images,values):
        c=c-1
        y.append(val.astype(int))
        X.append(img/255)
```

```python
y=np.matrix(y)
X=np.array(X)
```

```python
X_train, X_test, y_train, y_test = train_test_split( X, y, test_size=0.33, random_state=42)
```

```python
in_layer=Input(shape=[700,700,3])
conv_1 = Conv2D(32, (3, 3), activation="relu")(in_layer)
pool_1 = MaxPool2D()(conv_1)
conv_2 = Conv2D(64, (3, 3), activation="relu")(pool_1)
pool_2 = MaxPool2D()(conv_2)
conv_3 = Conv2D(128, (3, 3), activation="relu")(pool_2)
pool_3 = MaxPool2D()(conv_3)
conv_4 = Conv2D(256, (3, 3), activation="relu")(pool_3)
pool_4 = MaxPool2D()(conv_4)
conv_5 = Conv2D(512, (3, 3), activation="relu")(pool_4)
pool_5 = MaxPool2D()(conv_5)
conv_6 = Conv2D(1024, (3, 3), activation="relu")(pool_5)
pool_6 = MaxPool2D()(conv_6)
# conv_7 = Conv2D(2048, (3, 3), activation="relu")(pool_6)
# pool_7 = MaxPool2D()(conv_7)
# conv_8 = Conv2D(4096, (3, 3), activation="relu")(pool_7)
# pool_8 = MaxPool2D()(conv_8)
# conv_9 = Conv2D(8192, (3, 3), activation="relu")(pool_8)
# pool_9 = MaxPool2D()(conv_9)
# conv_10 = Conv2D(16388, (3, 3), activation="relu")(pool_9)
# pool_10 = MaxPool2D()(conv_10)
flatten = Flatten()(pool_6)
dense_1 = Dense(50000, activation="tanh")(flatten)
dense_2 = Dense(25000, activation="tanh")(dense_1)
dense_3 = Dense(10000, activation="tanh")(dense_2)
dense_4 = Dense(5000, activation="tanh")(dense_3)
# drop = Dropout(rate=.25)(dense_4)
# dense_5 = Dense(4000, activation="tanh")(drop)
# drop1 = Dropout(rate=.25)(dense_5)
# dense_6 = Dense(3000, activation="tanh")(drop1)
# drop2 = Dropout(rate=.25)(dense_6)
# dense_7 = Dense(1500, activation="tanh")(drop2)
# drop3 = Dropout(rate=.25)(dense_7)
# dense_8 = Dense(1000, activation="tanh")(drop3)
# drop4 = Dropout(rate=.25)(dense_7)
# dense_9 = Dense(900, activation="tanh")(drop4)
# drop5 = Dropout(rate=.25)(dense_8)
# dense_10 = Dense(450, activation="tanh")(drop5)
# drop6 = Dropout(rate=.25)(dense_9)
# dense_11 = Dense(250, activation="tanh")(drop6)
# drop7 = Dropout(rate=.25)(dense_10)
# dense_12 = Dense(150, activation="tanh")(drop7)
# drop8 = Dropout(rate=.25)(dense_11)
# dense_13 = Dense(100, activation="tanh")(drop8)
# drop9 = Dropout(rate=.25)(dense_12)
# dense_14 = Dense(50, activation="tanh")(drop9)
# drop10 = Dropout(rate=.25)(dense_13)
# dense_15 = Dense(25, activation="tanh")(drop10)
# drop11 = Dropout(rate=.25)(dense_14)
# dense_16 = Dense(15, activation="tanh")(drop11)
out = Dense(4, activation="softmax")(dense_4)
```

```python
model=Model(inputs=[in_layer],outputs=[out])
```

```python
model.summary()
```

```python
model.compile(optimizer="adam", loss="categorical_crossentropy", metrics=["acc"])
```

```python
model.fit(X_train,y_train,epochs=20,batch_size=2)
```
